```python
import torch
import numpy as np





```

```python
# auto infer data type
data = [[1, 2],[3, 4]]
x_data = torch.tensor(data)
print(x_data)
```

```python
# succeeds np arrays
np_array = np.array(data)
x_np = torch.from_numpy(np_array)
```

```python
# properties: shape, data type
x_ones = torch.ones_like(x_data) # only reads shape
print(f"Ones Tensor: \n {x_ones} \n")

# use dtype argument to override data type
x_rand = torch.rand_like(x_data, dtype=torch.float) # can't use rand_like on int matrix?
print(f"Random Tensor: \n {x_rand} \n")
```

```python
shape = (2,3,) # 2 row, 3 column
rand_tensor = torch.rand(shape) # randoms
ones_tensor = torch.ones(shape) # all 1
zeros_tensor = torch.zeros(shape) # all 0

# default datatype seems to be float?
print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
```

```python
tensor = torch.rand(3,4)

print(f"Shape of tensor: {tensor.shape}")
# default for rand is float32
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

```python
# We move our tensor to the GPU if available
# cuda refers to gpu?
if torch.cuda.is_available():
    tensor = tensor.to("cuda")
```

```python
tensor = torch.ones(4, 4)
print("First row: " + str(tensor[0])) # can force a 1d tensor of float32 to str type
print(f"First row: {tensor[0]}")
print(f"First column: {tensor[:, 0]}")
print(f"Last column: {tensor[..., -1]}")
tensor[:,1] = 0
print(tensor)
print("entire matrix: " + str(tensor)) # can force 2d tensor of float32 to str type as well
```

```python
t1 = torch.cat([tensor, tensor, tensor], dim=1) # concatenation, 0 based indexing for dimension
print(t1)
```

```python
# 3 different ways of matrix multiplication
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)

y3 = torch.rand_like(y1) # creating output storage tensor
torch.matmul(tensor, tensor.T, out=y3)



# * refers to element-wise multiplication, not matrix multiplication
z1 = tensor * tensor
z2 = tensor.mul(tensor)

z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)
```

```python
agg = tensor.sum() # sum of every element in the tensor (R^d -> 1), dtype is still tensor
print(agg)
agg_item = agg.item() # get the scalar item from tensor
print(agg_item, type(agg_item))
```

```python
print(f"{tensor} \n")
tensor.add_(5) # add 5 to every element, _ stores the change into original tensor
print(tensor)
```

```python
t = torch.ones(5)
print(f"t: {t}")
n = t.numpy() # 1d tensor into 1d numpy array representation, does not create new array
print(f"n: {n}")
```

```python
t.add_(1) # since tensor and np arrays stem from same memory, add 1 operation affects both
print(f"t: {t}")
print(f"n: {n}")
```

```python
n = np.ones(5)
t = torch.from_numpy(n)
print(f"n: {n}")
print(f"t: {t}")
```

```python
np.add(n, 1, out=n) # numpy does not have add_ operation
print(f"t: {t}")
print(f"n: {n}")
```
