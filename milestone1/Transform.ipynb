```python
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor, Lambda

# what does it do?
ds = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor(), # image into tensor
    # transformation of int into one-hot vector, creates zero array of size 10 and writes 1 at location torch.tensor(y)
    target_transform=Lambda(lambda y: torch.zeros(10, dtype=torch.float).scatter_(0, torch.tensor(y), value=1))
)
```
